```python
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('model_4.h5')
```

```python
model.summary()
```

```python
for layer in model.layers:
    print(f"Layer: {layer.name}")
    if hasattr(layer, 'weights'):
        weights = layer.weights
        for weight in weights:
            print(f"Weight shape: {weight.shape}")
            print(weight.numpy())  # Print or process the weights as needed
    if hasattr(layer, 'biases'):
        biases = layer.biases
        print(f"Bias shape: {biases.shape}")
        print(biases.numpy())  # Print or process the biases as needed

    # If you want to associate weights with input features:
    if isinstance(layer, tf.keras.layers.Dense):
        num_features = weights[0].shape[0]  # Assuming weights[0] are the kernel weights
        for i in range(num_features):
            print(f"Feature {i + 1} weight: {weights[0][i]}")
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
```

```python
with open('model_4.tflite', 'wb') as f:
    f.write(tflite_model)
```

```python
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_quantized_model = converter.convert()
```

```python
with open('model_4_quantized.tflite', 'wb') as f:
    f.write(tflite_quantized_model)
```
