```python
data=pd.read_csv('combined_features_data.csv')
```

```python
data['label']=data['label'].astype('int')
```

```python
data.reset_index(inplace=True)
data.drop(['index'], axis=1, inplace=True)
```

```python
import seaborn as sns
import os
```

```python
custom_palette = {
    0: 'orange',
    1: 'red',
    2: 'purple',
    3: 'green',
    4: 'blue',
    5: 'coral',
    6:'darkmagenta',
    7:'lightpink'
}
```

```python
# Define the label mapping
label_mapping = {
    0: 'CW_Circle',
    1: 'CCW_Circle',
    2: 'CW_Square',
    3: 'CCW_Square',
    4: 'CW_Triangle',
    5: 'CCW_Triangle',
    6: 'Double_Tap',
    7: 'S'
}

# Invert the mapping for easier access
label_mapping_inverted = {v: k for k, v in label_mapping.items()}
```

```python
save_dir = '/home/ettexham/Desktop/Thesis/Accelerometer Data/DataSet/Features/'

# Check if the directory exists, if not, create it
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
for feature in data.columns:
    if feature != 'label':
        # Create a filename for the plot based on the feature name
        filename = feature + '_boxplot.png'
        
        # Create the boxplot
        plt.figure(figsize=(10, 10))  # Set figure size
        sns.boxplot(data=data, x='label', y=feature, hue='label', palette='viridis')

        # Set labels and title
        plt.xlabel('Label')
        plt.ylabel(feature)
        plt.title(feature)
        
        # Customize x-axis ticks with label names
        plt.xticks(
            ticks=range(len(label_mapping)),  # Position for each label
            labels=[label_mapping[i] for i in range(len(label_mapping))],  # Map indices to names
            rotation=90,  # Rotate labels vertically
            ha='right'  # Align text to the right
        )
        
        # Save the plot to the specified folder
        plt.savefig(os.path.join(save_dir, filename))
        
        # Clear the current plot to prepare for the next iteration
        plt.clf()  # Clear the current figure

# Close any remaining plots
plt.close('all')
```
