```python
import pandas as pd
from dtw import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
#df.set_index('Time', inplace=True)
df=df[55:1960]
```

```python
plt.plot(df)
```

```python
timestamps = df['Time']
data_columns = df.drop(columns=['Time'])
```

```python
def dtw_similarity(series1, series2):
    # Calculate DTW distance with Euclidean distance
    distance = dtw(series1, series2, dist=lambda x, y: np.linalg.norm(x - y))
    return distance
```

```python
dtw_similarities = {}
for column1 in data_columns.columns:
    dtw_similarities[column1] = {}
    for column2 in data_columns.columns:
        if column1 != column2:
            series1 = data_columns[column1].values
            series2 = data_columns[column2].values
            dtw_similarities[column1][column2] = dtw_similarity(series1, series2)
```

```python
# Print DTW similarities
for column, similarities in dtw_similarities.items():
    print(f"DTW similarities for column '{column}':")
    for other_column, similarity in similarities.items():
        print(f"   {other_column}: {similarity}")
```

```python
def visualize_warping_path(series1, series2, path, column1, column2):
    import matplotlib.pyplot as plt

    plt.figure(figsize=(8, 6))
    plt.plot(series1, label=column1)
    plt.plot(series2, label=column2)
    plt.title('Comparison of Time Series')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()

    # Plot the warping path
    for i in range(len(path[0])):
        plt.plot([path[0][i], path[1][i]], [series1[path[0][i]], series2[path[1][i]]], color='red', linestyle='--')
```
