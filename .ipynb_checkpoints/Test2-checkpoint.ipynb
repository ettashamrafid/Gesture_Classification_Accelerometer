```python
def moving_average_filter(df, window_size):
    """
    Apply a moving average filter to the input DataFrame.

    Parameters:
        df (DataFrame): The input DataFrame.
        window_size (int): The size of the moving window.

    Returns:
        DataFrame: The filtered DataFrame.
    """
    # Apply the filter using rolling mean
    filtered_df = df.rolling(window=window_size, min_periods=1).mean()

    return filtered_df

window_size = 35
dff1 = moving_average_filter(df1[['x_data','y_data','z_data']], window_size)

# Print the results
print("Original DataFrame:")
print(df1)
print("\nFiltered DataFrame:")
print(dff1)
```

```python
fft_result = {}
for col in dff.columns:
    fft_values = np.fft.fft(dff[col])
    fft_freq = np.fft.fftfreq(len(dff), d=1/505)  # Compute the frequencies
    fft_magnitude = np.abs(fft_values) / len(dff)  # Compute the magnitudes
    fft_result[col] = (fft_freq, fft_magnitude)

# Plot the results
for col, (freq, mag) in fft_result.items():
    plt.plot(abs(freq), mag, label=col)

plt.title('FFT Results')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.signal import butter, filtfilt
sampling_rate=505

def low_pass_filter(data, cutoff_freq, fs):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(4, normal_cutoff, btype='low', analog=False)
    filtered_data = filtfilt(b, a, data)
    return filtered_data
```

```python
cutoff_frequency = 5  # Specify your cutoff frequency here
filtered_datax = low_pass_filter(dff['x_data'], cutoff_frequency, sampling_rate)
filtered_datay = low_pass_filter(dff['y_data'], cutoff_frequency, sampling_rate)
filtered_dataz = low_pass_filter(dff['z_data'], cutoff_frequency, sampling_rate)
```

```python
def calculate_statistics(df, column_name):
    column_data = df[column_name]
    stats_dict = {}
    
    stats_dict[column_name + '_min'] = np.min(column_data)
    
    stats_dict[column_name + '_max'] = np.max(column_data)

    stats_dict[column_name + '_range'] = np.ptp(column_data)

    # Calculate average
    stats_dict[column_name + '_avg'] = np.mean(column_data)

    # Calculate standard deviation
    stats_dict[column_name + '_std'] = np.std(column_data)

    # Calculate variance
    stats_dict[column_name + '_var'] = np.var(column_data)

    # Calculate root mean square (RMS)
    stats_dict[column_name + '_rms'] = np.sqrt(np.mean(column_data**2))

    return stats_dict
```

```python
def det_time(df, column_name):
    return np.max(df[column_name])- np.min(df[column_name])
```

```python
feature=pd.DataFrame()

dct_lst=[dff1,dff2,dff3,dff4,dff5,dff6]
for i in dct_lst:
    lst=['x_data', 'y_data', 'z_data', 'x_jerk', 'y_jerk', 'z_jerk', 'total', 'total_jerk']
    i['total'] = np.sqrt(i['x_data']**2 + i['y_data']**2 + i['z_data']**2)
    i['x_jerk'] = (i['x_data'] - i['x_data'].shift(1)) / (i['time'] - i['time'].shift(1))
    i['y_jerk'] = (i['y_data'] - i['y_data'].shift(1)) / (i['time'] - i['time'].shift(1))
    i['z_jerk'] = (i['z_data'] - i['z_data'].shift(1)) / (i['time'] - i['time'].shift(1))
    i['total_jerk'] = (i['total'] - i['total'].shift(1)) / (i['time'] - i['time'].shift(1))
    feature_dct={}
    feature_dct['time']=det_time(i, 'time')
    for j in lst:
        temp_dict= calculate_statistics(i, j)
        feature_dct.update(temp_dict)
    feature=feature.append(feature_dct, ignore_index=True)
```
