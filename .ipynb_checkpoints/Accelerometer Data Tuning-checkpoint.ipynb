```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
```

```python
df=df[1:]
df.set_index('Time', inplace=True)
```

```python
plt.plot(df)
```

```python
df.columns
```

```python
def apply_fft(column):
    fft_result = np.fft.fft(column)
    return fft_result
```

```python
fft_x=apply_fft(df['X'])
fft_y=apply_fft(df['Y'])
fft_z=apply_fft(df['Z'])
```

```python
sampling_rate = 10  # You need to specify your actual sampling rate here

# Number of data points
num_points = len(df)

# Frequencies corresponding to FFT result
frequencies = np.fft.fftfreq(num_points, d=1/sampling_rate)

# Plotting frequency vs amplitude curve
plt.figure(figsize=(10, 6))

# Plot for X-coordinate
plt.plot(frequencies, np.abs(fft_x), label='X-coordinate')
```

```python
# Plot for Y-coordinate
plt.plot(abs(frequencies), np.abs(fft_y), label='Y-coordinate')
```

```python
from scipy.signal import butter, filtfilt
```

```python
def low_pass_filter(data, cutoff_freq, fs):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(4, normal_cutoff, btype='low', analog=False)
    filtered_data = filtfilt(b, a, data)
    return filtered_data
```

```python
cutoff_frequency = 0.1  # Specify your cutoff frequency here
filtered_x = low_pass_filter(df['X'], cutoff_frequency, sampling_rate)
filtered_y = low_pass_filter(df['Y'], cutoff_frequency, sampling_rate)
filtered_z = low_pass_filter(df['Z'], cutoff_frequency, sampling_rate)
```

```python
plt.plot(filtered_x)
```

```python
plt.plot(filtered_y)
```

```python
plt.plot(filtered_z)
```

```python
data={
    'X':filtered_x,
    'Y':filtered_y,
    'Z':filtered_z
}

filtered_df=pd.DataFrame(data)
```
