```python
def moving_average_filter(df, window_size):
    """
    Apply a moving average filter to the input DataFrame.

    Parameters:
        df (DataFrame): The input DataFrame.
        window_size (int): The size of the moving window.

    Returns:
        DataFrame: The filtered DataFrame.
    """
    # Apply the filter using rolling mean
    filtered_df = df.rolling(window=window_size, min_periods=1).mean()

    return filtered_df

window_size = 15
df1 = moving_average_filter(df[['X','Y','Z']], window_size)

# Print the results
print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame:")
print(df1)
```

```python
df_t=df.drop(['Time'], axis=1)
```

```python
fft_result = {}
for col in df_t.columns:
    fft_values = np.fft.fft(df_t[col])
    fft_freq = np.fft.fftfreq(len(df_t), d=1/9.9)  # Compute the frequencies
    fft_magnitude = np.abs(fft_values) / len(df_t)  # Compute the magnitudes
    fft_result[col] = (fft_freq, fft_magnitude)

# Plot the results
for col, (freq, mag) in fft_result.items():
    plt.plot(abs(freq), mag, label=col)

plt.title('FFT Results')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.signal import butter, filtfilt
sampling_rate=9.9

def low_pass_filter(data, cutoff_freq, fs):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(4, normal_cutoff, btype='low', analog=False)
    filtered_data = filtfilt(b, a, data)
    return filtered_data
```

```python
cutoff_frequency = 1.5  # Specify your cutoff frequency here
filtered_datax = low_pass_filter(df['X'], cutoff_frequency, sampling_rate)
filtered_datay = low_pass_filter(df['Y'], cutoff_frequency, sampling_rate)
filtered_dataz = low_pass_filter(df['Z'], cutoff_frequency, sampling_rate)
```

```python
type(filtered_datax)
```

```python
data = {'x_data': filtered_datax , 'y_data': filtered_datay, 'z_data': filtered_dataz }

filtered_datam=pd.DataFrame(data)
```
