```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('feature.csv')
```

```python
data.drop(['x_jerk_range', 'y_jerk_range', 'z_jerk_range', 'total_jerk_range'], axis=1, inplace=True)
data['folder_label']=data['folder_label'].astype(int)
```

```python
data['folder_label']-=1
```

```python
data
```

```python
import seaborn as sns
import os
```

```python
custom_palette = {0: 'orange', 1: 'blue', 2: 'red'}

# Assuming you already have your dataframe, let's call it data

# Set the directory where you want to save the plots
save_dir = '/home/ettexham/Desktop/Thesis/Accelerometer Data/'

# Check if the directory exists, if not, create it
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# Iterate through each column in the dataframe
for column in data.columns:
    # Create a filename for the plot based on the column name
    filename = column + '.png'
    
    # Create the plot
    sns.lineplot(data=data, x=data.index, y=column, hue='folder_label', palette=custom_palette)
    
    # Set labels and title
    plt.xlabel('Index')
    plt.ylabel(column)
    plt.title(column)
    
    # Save the plot to the specified folder
    plt.savefig(os.path.join(save_dir, filename))
    
    # Clear the current plot to prepare for the next iteration
    plt.clf()  # Clear the current figure
    
# Close any remaining plots
plt.close('all')
```
